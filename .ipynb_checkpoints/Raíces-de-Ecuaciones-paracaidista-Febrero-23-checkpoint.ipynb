```python
import math
import numpy as np
import pylab as pl

valoresx = []
valoresy = []
valoresejex=[]
valoresejey=[]
valoresx1=[]
valoresy1=[]

g=9.8
m=68.1
t=10
v=40
xl=12.0
xu=16.0

print '{0} \t {1} \t {2} \n '.format("xl","xu","xr")

def f(c):
    return ((g*m/c)*(1-math.exp(-(c/m)*t)))-v

def fxr(xl,xu):
    return (xl+xu)/2
if((f(xl)*f(xu))<0):
    xr=fxr(xl,xu)

    print '{0} \t {1} \t {2} '.format(xl,xu,xr)
    
while((f(xl)*f(xr))!=0):
    if((f(xl)*f(xr))<0):
        xu=xr
        xr=fxr(xl,xu)
        print '{0} \t {1} \t {2} '.format(xl,xu,xr)
        
    else:
        if((f(xl)*f(xr))>0):
            xl=xr
            xr=fxr(xl,xu)
            print '{0} \t {1} \t {2} '.format(xl,xu,xr)

x=12        
i=0

while(i<=20):
    valoresejex.append(i)
    valoresejey.append(0)
    i+=1
while(x<=16):
    valoresx.append(x)
    valoresy.append(f(x))
    x=x+0.01
    
valoresx1.append(0)   
valoresy1.append(xr)   
            
f1=  pl.plot(valoresx,valoresy,"r")
f2=  pl.plot(valoresejex,valoresejey,"b")
f3=  pl.plot(valoresy1,valoresx1,".")


pl.title("Paracaidista \n")
pl.xlabel(" C \n")
pl.ylabel(" F(c) ")
f=[" Funcion "]
pl.legend(f,loc=4)
pl.xlim(10,17)
pl.ylim(-4, 7)
pl.grid(True)

pl.show()
```
