```python
import math
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2
%matplotlib inline 

x=[ 6, 7, 11, 15, 17, 21, 23, 29, 29, 37, 39] 
y = [29, 21, 29, 14, 21, 15, 7, 7, 13, 0, 3]
x2=[ 6, 7, 11, 15, 17, 21, 23, 29, 29, 37, 39,10] 
y2= [29, 21, 29, 14, 21, 15, 7, 7, 13, 0, 3,10]

tiempo=[]
ecuacion1=[]
ecuacion2=[]
valores=[]
n=len(x)
n2=len(x2)

for t in range (1, 16):
    tiempo.append(t)

def sumaProd(vec1,vec2, cont):
    suma1=0
    for i in range(cont):
        suma1+=(vec1[i]*vec2[i])
    return suma1

def sumaXY(vec1,vec2,cont):
    suma2=0
    suma3=0
    for j in range (cont):
        suma2+=vec1[j]
        suma3+=vec2[j]
    producto=suma2*suma3
    return producto

def sumaCuadr(vec1, cont):
    suma4=0
    for k in range(cont):
        suma4+=math.pow(vec1[k],2)
    return suma4

def cuadrado(vec1, cont):    
    suma5=0
    for l in range(cont):
        suma5+=vec1[l]
    return (math.pow(suma5,2))

def mediaX(vec1, cont): 
    suma6=0
    for o in range(cont):
        suma6+=vec1[o]
    return (suma6/len(vec1))

def mediaY(vec2, cont):
    suma7=0
    for m in range(cont):
        suma7+=vec2[m]
    return (suma7/len(vec2))

def raices(vec1,vec2,cont):
    raiz1=0
    raiz2=0
    sumax=0
    sumacx=0
    sumay=0
    sumacy=0
    for k in range(cont):
        sumax+=math.pow(vec1[k],2)
        sumacx+=vec1[k]
        sumay+=math.pow(vec2[k],2)
        sumacy+=vec2[k]
    raiz1=math.sqrt(((cont*sumax)-(math.pow(sumacx,2))))
    raiz2=math.sqrt(((cont*sumay)-(math.pow(sumacy,2))))
    return (raiz1*raiz2)

def coeficiente(vec1,vec2,cont):
    r=(((cont*sumaProd(vec1,vec2,cont))-sumaXY(vec1,vec2,cont))/(raices(vec1,vec2,cont)))
    return (r)

def error(vec1,vec2,a1,a0,cont):
    sr=0
    for p in range(cont):
        sr+=math.pow(vec2[p]-a0-a1*vec1[p],2)
    syx=math.sqrt(sr/(cont-2))
    return syx
    
a1A=(n*sumaProd(x,y,n)-sumaXY(x,y,n))/(n*sumaCuadr(x,n)-cuadrado(x,n))
a0A=mediaY(y,n)-a1A*mediaX(x,n)

a1B=(n2*sumaProd(x2,y2,n2)-sumaXY(x2,y2,n2))/(n2*sumaCuadr(x2,n2)-cuadrado(x2,n2))
a0B=mediaY(y2,n2)-a1B*mediaX(x2,n2)

for w in range (40):
    func1=a0A+(a1A*w)
    ecuacion1.append(func1)
    func2=a0B+(a1B*w)
    ecuacion2.append(func2)
    valores.append(w)
    
print "Para la primera parte:\n "
print "El coeficiente de correlacion es = ", coeficiente(x,y,n)
print "El error estandar estimado es = ", error(x,y,a1A,a0A,n)

plt.title("Grafica #1")
plt.grid(True)
plt.axis([0,40,0,30])
plt.ylabel('f(x)')
plt.xlabel('x')
plt.plot(x, y, 'co', label="Valores iniciales")
plt.plot(valores, ecuacion1, 'r--', linewidth=2.0, label="y=30.391-0.780x")
plt.legend(loc = 'lower left', numpoints = 2)
plt.show()

print "Agregando x=10 , y=10:\n "
print "El coeficiente de correlacion es = ", coeficiente(x2,y2,n2)
print "El error estandar estimado es = ", error(x2,y2,a1B,a0B,n2)

plt2.title("Grafica #2")
plt2.grid(True)
plt2.axis([0,40,0,30])
plt2.ylabel('f(x)')
plt2.xlabel('x')
plt2.plot(x2, y2, 'ro', label="Valores iniciales")
plt2.plot(valores, ecuacion2, 'b--', linewidth=2.0, label="y=27.655-0.683x")
plt2.legend(loc = 'lower left', numpoints = 2)
plt2.show()

print "La medición es invalida, se puede observar en la gráfica #2 que se \ncomporta como valor atipico ya que esta muy distante\
 del resto de valores. \nAdemás, agregando x=10, y=10, el error estándar cuantifica una \nmayor dispersión de los datos \
 alrededor de la línea de regresión."
```
