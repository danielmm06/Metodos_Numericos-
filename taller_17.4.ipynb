```python
import math
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2
%matplotlib inline 

x = [0, 2, 4, 6, 9, 11, 12, 15, 17, 19]  
y = [5, 6, 7, 6, 9, 8, 7, 10, 12, 12]

tiempo=[]
ecuacion1=[]
ecuacion2=[]
valores=[]
n=len(x)

for t in range (1, 16):
    tiempo.append(t)

def sumaProd(vec1,vec2):
    suma1=0
    for i in range(n):
        suma1+=(vec1[i]*vec2[i])
    return suma1

def sumaXY(vec1,vec2):
    suma2=0
    suma3=0
    for j in range (n):
        suma2+=vec1[j]
        suma3+=vec2[j]
    producto=suma2*suma3
    return producto

def sumaCuadr(vec1):
    suma4=0
    for k in range(n):
        suma4+=math.pow(vec1[k],2)
    return suma4

def cuadrado(vec1):    
    suma5=0
    for l in range(n):
        suma5+=vec1[l]
    return (math.pow(suma5,2))

def mediaX(vec1): 
    suma6=0
    for o in range(n):
        suma6+=vec1[o]
    return (suma6/len(vec1))

def mediaY(vec2):
    suma7=0
    for m in range(n):
        suma7+=vec2[m]
    return (suma7/len(vec2))

def raices(vec1,vec2):
    raiz1=0
    raiz2=0
    sumax=0
    sumacx=0
    sumay=0
    sumacy=0
    for k in range(n):
        sumax+=math.pow(vec1[k],2)
        sumacx+=vec1[k]
        sumay+=math.pow(vec2[k],2)
        sumacy+=vec2[k]
    raiz1=math.sqrt(((n*sumax)-(math.pow(sumacx,2))))
    raiz2=math.sqrt(((n*sumay)-(math.pow(sumacy,2))))
    return (raiz1*raiz2)

def coeficiente(vec1,vec2):
    r=(((n*sumaProd(vec1,vec2))-sumaXY(vec1,vec2))/(raices(vec1,vec2)))
    return (r)

def error(vec1,vec2,a1,a0):
    sr=0
    for p in range(n):
        sr+=math.pow(vec2[p]-a0-a1*vec1[p],2)
    syx=math.sqrt(sr/(n-2))
    return syx
    
a1A=(n*sumaProd(x,y)-sumaXY(x,y))/(n*sumaCuadr(x)-cuadrado(x))
a0A=mediaY(y)-a1A*mediaX(x)

a1B=(n*sumaProd(y,x)-sumaXY(y,x))/(n*sumaCuadr(y)-cuadrado(y))
a0B=mediaY(x)-a1B*mediaX(y)

for w in range (20):
    func1=a0A+(a1A*w)
    ecuacion1.append(func1)
    func2=a0B+(a1B*w)
    ecuacion2.append(func2)
    valores.append(w)
    
print "Para la primera parte:\n "
print "La pendiente o coeficiente a1 = ", a1A
print "La interseccion eje Y o coeficiente a0 = ", a0A
print "El coeficiente de correlacion es = ", coeficiente(x,y)
print "El error estandar estimado es = ", error(x,y,a1A,a0A)

plt.title("Grafica 1")
plt.grid(True)
plt.axis([0,20,0,15])
plt.ylabel('F(X))')
plt.xlabel('X')
plt.plot(x, y, 'co', label="Valores iniciales")
plt.plot(valores, ecuacion1, 'r--', linewidth=2.0, label="y=4.823+0.352x")
plt.legend(loc = 'lower right', numpoints = 2)
plt.show()

print "Para la segunda parte:\n "
print "La pendiente o coeficiente a1 = ", a1B
print "La interseccion eje Y o coeficiente a0 = ", a0B
print "El coeficiente de correlacion es = ", coeficiente(y,x)
print "El error estandar destimado es = ", error(y,x,a1B,a0B)

plt2.title("Grafica 2")
plt2.grid(True)
plt2.axis([0,15,-10,25])
plt2.ylabel('F(X)')
plt2.xlabel('x')
plt2.plot(y, x, 'ro', label="Valores iniciales")
plt2.plot(valores, ecuacion2, 'b--', linewidth=2.0, label="y=-9.992+2.374x")
plt2.legend(loc = 'lower right', numpoints = 2)
plt2.show()

print "Se puede observar que:\n\n 1) El coeficiente de correlacion es el mismo para ambas partes. Esto quiere decir que así\
 se alternen\n    las variables, su relación sigue siendo igual. En estos casos, el coeficiente se aproxima a 1,\n   \
 lo que significa que las relaciones son fuertes y directas.\n\n 2) Para la primera parte se obtiene un mejor ajuste por \
 minimos cuadrados. Pues la recta esta\n    mas cercana a los puntos dados inicialmente."
```
