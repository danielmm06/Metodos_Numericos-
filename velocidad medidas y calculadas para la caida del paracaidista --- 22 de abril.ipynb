```python
import pylab as plt
import numpy as np
import math

med=[10,16.3,23,27.5,31,35.6,39,41.5,42.9,45,46,45.5,46,49,50]
B=[]
A_1=[]

g=9.8
c=12.5
m=68.1
n = 15

print'|{0:7} |  {1:10} | {2:20} | {3:18} |'.format(' tiempo','Valor medido','valor calculado 1.10','valor calculado 17')

for t in range(n):
    v=((g*m)/c)*(1-math.exp((-c/m)*(t+1)))
    v_1=((g*m)/c)*((t+1)/(3.75+(t+1)))
    
    print'|{0:7} |  {1:12} | {2:20} | {3:18} |'.format(t+1,med[t],round(v,3),round(v_1,3))
    
    A_1.append([])
    for j in range(1):
        A_1[t].append(t)
        A_1[t].append(v)
        
A=np.array(A_1)
lo=0
li_2=0
x=[]
x_2=0
vx=[]
vy=[]

print "\n\nRepresentacion en Matriz:\n"
print A, "\n"

for l in range(2):
    m=(A.size)/2
    for k in range(m):
        lo=A[k,l]+lo
        if l==0:
            li_2=(A[k,l]**2)+li_2
        if l==1:
            x_2=A[k,0]*A[k,1]+x_2
        if k==(m-1):
            x.append(lo)
            
            if l==1:
                x.append(x_2)
                x.append(li_2)
            lo=0
print x

a_2=((m*x[3])-(x[0])**2)
a_1=((m*x[2])-(x[0]*x[1]))/((m*x[3])-(x[0])**2)
a_0=(x[1]/m)-a_1*(x[0]/m)

print a_1
print a_0

plt.title("Grafica A \n")
plt.grid(True)
plt.axis([-20,20,0,60])
plt.ylabel('X_2')
plt.xlabel('X_1')

for k in range(m):
    y=a_0+a_1*k
    vx.append(k)
    vy.append(y)
    plt.plot(A[k,0],A[k,1], 'cs')
plt.plot(vx,vy,'r')
plt.show()
```
