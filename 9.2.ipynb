```python
A=[[4,7],[1,2],[5,6]]
B=[[4,3,7],[1,2,7],[1,0,4]]
C=[[3],[6],[1]]
D=[[9,4,3,-6],[2,-1,7,5]]
E=[[1,5,8],[7,2,3],[4,0,6]]
F=[[3,0,1],[1,7,3]]
G=[7,6,4]
I=[[1,0,0],[0,1,0],[0,0,1]]

#1) suma de matrices [E]+[B]
sumaEB=[]
for i in range(3): # filas
    sumaEB.append([])
    for j in range(3):#  columnas
        sumaEB[i].append(E[i][j]+B[i][j])

print "1) La suma de las matrices E + B es: "
print sumaEB

#2) suma de matrices [A]+[F]
print "\n2) La suma de las matrices A + F no se puede realizar porque no tienen la misma dimension m x n "

#3) resta de as matrices [B]-[E]
restaBE=[]
for i in range(3):
    restaBE.append([])
    for j in range(3):
         restaBE[i].append(B[i][j]-E[i][j])

print "\n3) La resta de las matrices B - E es: "
print restaBE

#4)miltiplacion del escalar 7 por la matriz B ---> 7 x [B]
escalar=[]
for i in range(3):
    escalar.append([])
    for j in range(3):
         escalar[i].append(7*B[i][j])

print "\n4) Al multiplicar [B] por el escalar 7, se obtiene: "
print escalar

def multiplicacion(a,b):
    product=[]
    for i in range(len(a)):
        product.append([0]*(len(b[0])))
    for i in range(len(a)):
        for j in range(len(b[0])):
            for k in range(len(a[0])):
                product[i][j]+= a[i][k]*b[k][j]
    return product

def transpuesta(a):
    trans=[]
    for i in range(len(a[0])):
        trans.append(len(a)*[0])
        for j in range(len(trans[0])):
            for k in range(len(trans)):
                trans[k][j]=a[j][k]
    return trans
#5 producto cruz entre [E]x[B]
print "\n5)El producto entre las matrices E x B es: "
print multiplicacion(E,B)

#6){C}^T
print "\n6) La transpuesta de {C} es: "
print transpuesta(C)

#7) producto cruz entre [B]x[A]
print "\n7)El producto entre las matrices B x A es: "
print multiplicacion(B,A)

#8)[D]^T
print "\n8)La transpuesta de [D] es: "        
print transpuesta(D)

#9) el procto cruz entre [A]x{C}
print "\n9) El producto [A] x [C] no se puede realizar porque el numero de columnas de A, \nNO coincide con el numero de filas de C"

#10) producto cruz entre [F]x[B]
print "\n10) El producto de [I] x [B] es: "            
print multiplicacion(I,B)

#11)[E]^T x [E] --> transpuesta de la matriz E por la matriz E           
print "\n11) El producto de [E]^T x [E] es: "
print multiplicacion(transpuesta(E),E)

#12) {C}^T x {C}--> producto de la transpuesta de la matriz C por la matriz C
print "\n12) El producto de {C}^T x {C} es: "
print multiplicacion(transpuesta(C),C)
```
